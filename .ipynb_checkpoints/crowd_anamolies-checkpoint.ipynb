```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python



audio_file = 'separated_audio/Crystal_Palace_VS_Man_City_crowd.wav'
y, sr = librosa.load(audio_file, sr=None, mono=True)  


frame_length = 2048
hop_length = 512
rms = librosa.feature.rms(y=y, frame_length=frame_length, hop_length=hop_length)[0]


frames = range(len(rms))
t = librosa.frames_to_time(frames, sr=sr, hop_length=hop_length)


plt.figure(figsize=(10, 6))
plt.plot(t, rms, label="RMS Energy")
plt.title("Crowd Noise Energy")
plt.xlabel("Time (s)")
plt.ylabel("Energy")
plt.legend()
plt.show()


threshold = np.mean(rms) + 3.4 * np.std(rms)
anomalies = t[rms > threshold]

print(f"Detected anomalies at times (seconds): {anomalies}")
```
